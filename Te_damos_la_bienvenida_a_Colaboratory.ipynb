H0: Xa = Xb Hipotesis nula

H1: Xa != Xb — Hipotesis Alterna

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
sample_a = np.random.normal(24, 10, 100)
mean(sample_a)

sample_b = np.random.normal(24, 10, 100)
mean(sample_b)
plt.figure(figsize=(20, 10))
sns.kdeplot(sample_a, label="A", shade = True, color = 'LIME')
sns.kdeplot(sample_b, label="B", shade = True, color = 'YELLOW')
plt.axvline(x=mean(sample_a),ymin=0, ymax=1)
plt.axvline(x=mean(sample_b),ymin=0, ymax=1,color ="r")
plt.legend();
```

```
import statsmodels.api as sm
import scipy.stats as stats
## H0: A = B
## H1: A != B
test = stats.ttest_ind(sample_a,sample_b, equal_var = True)
print(test)


```

```
statistic, pvalue = test
print('statistic=%.3f, pvalue=%.3f' % (statistic, pvalue))
if pvalue > 0.05:
	print('NO Rechazo Ho')
else:
	print('Se acepta Ha')
```
